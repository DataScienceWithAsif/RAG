```python
import re
from collections import Counter
import math

def cosine_similarity_func(query, doc):
    # clean and tokenize (remove punctuation, lowercase, split by spaces)
    def tokenize(text):
        text = re.sub(r'[^a-zA-Z0-9 ]', '', text)  # remove punctuation
        return text.lower().split()

    query_tokens = tokenize(query)
    doc_tokens = tokenize(doc)

    query_counter = Counter(query_tokens)
    doc_counter = Counter(doc_tokens)

    # dot product
    dot_product = sum(query_counter[t] * doc_counter[t] for t in query_counter if t in doc_counter)

    # magnitudes
    query_magnitude = math.sqrt(sum(v**2 for v in query_counter.values()))
    doc_magnitude = math.sqrt(sum(v**2 for v in doc_counter.values()))

    if query_magnitude == 0 or doc_magnitude == 0:
        return 0.0

    return dot_product / (query_magnitude * doc_magnitude)
```

```python
cosine_similarity=cosine_similarity_func("high-level programming language","Python is a high-level programming language. It is widely used for data analysis, machine learning, web development, and automation.")
cosine_similarity
```

```python
cosine_similarity=cosine_similarity_func("high-level programming language","Deep Learning is a specialized form of Machine Learning using neural networks with many layers. It is widely applied in image recognition, speech processing, and text generation.")
cosine_similarity
```

```python
def return_response(query, documents):
    similarities=[]
    for doc in documents:
        similarity=cosine_similarity_func(query,doc)
        similarities.append(similarity)

    return documents[similarities.index(max(similarities))]
```
